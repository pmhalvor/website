# Neural Networks

This note will attempt to demystify neural networks. I'll explain the basic theory of neural networks, then follow up with a simple implementation written in Python. 


# Plainly explained

You have a dataset with a bunch of _features_. Using the rest of the features in the dataset, you want to predict a single feature, call it the _target._ 

Some of these features may heavily affect the accuracy of predicting our target. Others may have much less influence. We can give each feature a weight, $w_i$, based off this _correlation._ 

The vector representing weights $\vec{w}=[w_1, ..., w_n]$ multiplied by the value of each feature can then be summed, represented as $z$, and funneled into a sigmoid function, $\sigma (z)$, below represented as $\hat y$. 


> _The 'bias node' is an extra parameter representing the intercept of the function. This node is not dependent on a single feature, but rather of the collection of all the features._

The sigmoid function is one among many activation functions which can be used to predict the value of a particular target.
