```r
data.leuk = read.csv("http://web.stanford.edu/~hastie/CASI_files/DATA/leukemia_big.csv") # direct url for up to date data

X <- data.leuk

# Split data according to obersvations mentioned above
test.obs <- colnames(X) %in% c('ALL.4', 'ALL.8', 'ALL.10', 
                   'ALL.11', 'ALL.13', 'ALL.18', 
                   'AML', 'AML.1', 'AML.4', 'AML.6', 
                   'AML.8', 'ALL.23', 'ALL.26',
                   'ALL.29', 'ALL.31', 'ALL.32', 
                   'ALL.35', 'ALL.39', 'ALL.40', 
                   'ALL.41', 'ALL.42', 'AML.16', 
                   'AML.22', 'AML.24')


X.test <- t(X[test.obs ])
y.test <- rep(0, nrow(X.test))
y.test[grepl("ALL", rownames(X.test))] <- 1 # ALL will have value 1, AML will have value 0

X.train <- t(X[!test.obs])
y.train <- rep(0, nrow(X.train))
y.train[grepl("ALL", rownames(X.train))] <- 1 # ALL will have value 1, AML will have value 0

head(X.train)
head(X.test)
```

```r
library(glmnet)

lambda <- exp(c(-7:2))

# lasso
lasso.l <- cv.glmnet(y = y.train, x = as.matrix(X.train),  lambda=lambda, alpha=1, nfold=71, standardize = TRUE, family='binomial')
lasso <- glmnet(y=y.train, x = as.matrix(X.train), lambda=lasso.l$lambda.min, alpha=1, standardize=TRUE, family='binomial')

#ridge
ridge.l <- cv.glmnet(y = y.train, x = as.matrix(X.train), lambda=lambda, alpha = 0, nfold=71, standardize = TRUE, family='binomial')
ridge  <- glmnet(y = y.train, x = as.matrix(X.train), lambda=ridge.l$lambda.min, alpha = 0, standardize = TRUE, family='binomial')
```

```r
lasso.l$lambda.min
ridge.l$lambda.min
```

In both of these methods,  $\lambda$ is used to help shrink the model relative to the intercepts. So now, instead of comparing only the least squares estimators between models, we add some bias to the model, to hopefully lower the variance even further. 

For lasso, $\lambda$ does this by limiting how many relevant variables should be included in the model. As $\lambda$ decreases, the number of variables included in the lasso model increases. 

For ridge regression, $\lambda$ weighs the coefficients depending on their relevance to the model. Again, as $\lambda$ decreases, the amount by which the less relevant variables affect the model increases. 

For both methods, $\lambda=0$ gives the regular least squares estimator, while $\lambda \rightarrow \infty$ gives an increasingly larger MSE. Somewhere between these two extremes, however, lies an optimal $\lambda$ which gives the lowest MSE for each of these respective model, which can be found using cross validation.

After executing a LOOCV for both models, we found that an $i=4$ gave $\lambda \approx 0.0183$ for both lasso and ridge. This was used to create the optimal lasso and ridge models above.

```r
# Predict mean square prediction error for lasso and ridge
train.error <- list()
train.error$ridge <- mean((y.train - ifelse(predict(ridge, newx = X.train, type = 'response') > 0.5, 1, 0))^2)
train.error$lasso <- mean((y.train - ifelse(predict(lasso, newx = X.train, type = 'response') > 0.5, 1, 0))^2)

test.error <- list()
test.error$ridge <- mean((y.test - ifelse(predict(ridge, newx = X.test, type = 'response') > 0.5, 1, 0))^2)
test.error$lasso <- mean((y.test - ifelse(predict(lasso, newx = X.test, type = 'response') > 0.5, 1, 0))^2)

cbind(train.error, test.error)
    
```

We got 0 for our misclassification error for lasso test and train, as well as for ridge training. It was only for ridge test that we got  `test.error.ridge = 0.041667`$>0$. In other words, on our test data, the ridge model misclassified only a single observation. 

The training errors should usually be smaller than the test errors since it is precisely these training data that are used to _train_ the model. This means that the model should fit these data better than any new data coming in. 

Since we were told to use a deterministic cross-validation, we needed to use the amount of folds that will give the same answer every time, no matter what random seed is set, namely the "leave-one-out" cross validation. We therefore set our number of folds equal to $(n-1)=71$, where $n=72$ is the number of observations we have. 

The advantage with using LOOCV is that it gives the best possible estimate of mean square error. This is because it uses every piece of data in the set as a test set on a model built with all the remaining data points, thus getting the absolute most out of the training data. 

The disadvantage is an increase in computation time. With _very_ large data sets, LOOCV can cost too must processing power to be able to compute in a reasonable amount of time. 

A lift curve provides a measure of the improvement of classification of a model with respect to classification by chance. 

The algorithm which generates a lift curve sorts the observations according to how likely each data point is of being classified correctly. The points most likely to be _classified correctly_ will come first in our fractions of predicted subjects (the left of the x-axis), and the observations that are most likely to be _misclassified_ will be pushed toward the back (toward the right of the x-axis).

Even though the test data shows there is a slight chance for misclassification of cancer type when using the ridge model, this model can still produce the same lift curve as the lasso model. This is because the observations most likely to be correctly classified are prioritized over those most likely to be misclassified in lift curves. In other words, lift curves focus on the success of a model, while slightly blurring the model's downfalls. 


The point at $(.25, 3)$ shows the ideal lift of both models. This point tells us that the top $25\%$ of our sorted training set contains $3\cdot 25\% = 75\%$ of our correctly classified data points. Another way of interpreting it is to say that looking at the $25\%$ of the data set which is easiest to classify, the model is 3 times better at classifying than if we had randomly chosen classifications for the observations. As we move along the x-axis, increasing the fraction of predicted subjects, the ratio of correctly classified observations vs misclassified is compared to the ratio between the two cancer types.

The misclassification found for this procedure is 0. In the next exercise, we'll look into why this cannot actually be compared to the lasso and ridge models from above.

The above procedure is wrong since we are using the entire data set to find the best 9 genes, and _then_ training on part of the data set and using the other part to test. 

If we want to fix the procedure, we must split the data set into training and test sets _first,_  and after that use _only the training set_ to find the 9 best genes.  

That way our test data is completely new data for our model, and will give a true estimate of how well our model is built. 


```r
### Step 1: Split the data into training/test data
# reread in X
X <- t(data.leuk)
colnames(X) <- paste('gene', c(1:nrow(data.leuk)))

#normalize
normalize <- function(x) {return ((x - min(x)) / (max(x) - min(x)))}
X <- normalize(X)


#Split the data set acoording to exercise a)

X.train <- X[!test.obs,]
y.train <- rep(0, nrow(X.train))
y.train[grepl("ALL", rownames(X.train))] <- 1 # ALL will have value 1, AML will have value 0




### Step 2: Find the 9 best genes using a t-test
# Split the two cancer types 
X.ALL <- X.train[grepl("ALL", rownames(X.train)),]
X.AML <-X.train[!grepl("ALL", rownames(X.train)),]

#t-test
res.est <- numeric(7128)
for(i in 1:7128){
    res <- t.test(x=X.ALL[,i], y=X.AML[,i])$estimate[1:2]
    res.est[i] <- abs(res[1]-res[2])
}

res.sorted <- sort(res.est, index.return=TRUE)
best.genes <- rev(tail(res.sorted$ix, n=9))
best.genes


# Isolate the best genes
X <- t(data.leuk)
colnames(X) <- c(1:nrow(data.leuk))
best <- colnames(X) %in% best.genes
X.best <- X[,best]

#split according to same as above
X.test <- X.best[test.obs ,]
y.test <- rep(0, nrow(X.test))
y.test[grepl("ALL", rownames(X.test))] <- 1 # ALL will have value 1, AML will have value 0

X.train <- X.best[!test.obs,]
y.train <- rep(0, nrow(X.train))
y.train[grepl("ALL", rownames(X.train))] <- 1 # ALL will have value 1, AML will have value 0




### Step 3 (same)
#Fit log.regression on train set
logistic <- glm(y.train ~ . , data=data.frame(X.train), family = 'binomial')


# logistic
# due to numerical error, not include all variables
y.hat.logistic <- ifelse(predict(logistic, newdata = data.frame(X.test), type = 'response') > 0.5, 1, 0)



### Step 4 (same)
# find error on test set
test.error$logistic <- mean((y.test - y.hat.logistic)^2)
test.error$logistic
```

We still get 0 as our misclassification error, but at least now our model was built using the same restrictions as our models in a. 

```r
# Merge data sets
X <- t(data.leuk)
colnames(X) <- paste('gene',c(1:nrow(data.leuk)))

# Select initialization points
init.rows <- rownames(X) %in% c("ALL", "ALL.10", "ALL.11", "AML", "AML.2")
init <- X[init.rows,]

# create y column
y <- rep(0, nrow(X))
y[grepl("ALL", rownames(X))] <- 1


# compute kmeans on our data
for(i in 1:4){
    for(j in (i+1):5){
        km.res <- kmeans(X, init[c(i,j), ])
        label <- km.res$cluster-1
        y[grepl("ALL", rownames(X))] <- label["ALL"]
        y[grepl("AML", rownames(X))] <- label["AML"]        
        
        print(sum(abs(y-label))/72)
    }
}
```

The difference between the different combinations of observations tells us that some are better fits for initialization points than others. The observations giving the higher errors should not be used as initialization points, since they are more likely outliers than close to the core of the dataset's actual clusters. Instead, the combinations of observations giving 0.01389 as the misclassification error should be used for initialization points. 


For this dataset, we already know we are looking for two different clusters, so choosing $K=2$ is straightforward. For datasets where the structure is unknown, finding the correct K can be pretty tricky.

In general, the K-mean algorithm can be problematic since it will only find clusters of elliptical shapes. Also, setting an arbitrary value for K will always give exactly that many clusters, even though the data set may contain more or less. In other words, the algorithm might split up a cluster just to satisfy the requirement for K, while disregarding the actual clustering of the dataset. This means we need to somehow find some optimal K for the data we are observing. 


Increasing the number of K's will always give a lower deviance in the clusters. Taking this to the furthest extreme, a K equal to the number of observations will give a deviance within the clusters equal to zero, $D_{within}=0$, since each cluster will contain only one data point. While we _are_ looking to minimize $D_{within}$, we need to somehow limit how large our K is, so that we're not breaking actual clusters into smaller sub-clusters.

If a data set has for example 4 clusters, and we increase some K$<4$ by 1, we will observe a huge drop in $D_{within}$. However, if we create a new cluster when there are no more there (i.e. increase K by 1 when we're already at optimal K), the deviance will still drop, but only by a small amount.  These small drops in deviance will still be seen as improvements by a cross validation, even though it does not give the better value for K. We want K to be as small as possible with still giving a small $D_{within}$. We are therefore looking for the elbow case.

To find an optimal K, we can calculate $D_{within}$ for different K's, and plot the average deviances. K can then be chosen by finding the elbow-case, or the value of K where the relative change in $D_{within}$ begins to flatten out. 


```r
# reread in X
X <- t(data.leuk)

# find distances
d <- dist(X)

# create heir. cluster
hier.cluster <- hclust(d, method = "complete")
plot(hier.cluster, sub = "", lwd = 2)
```

Even though there seems to be a grouping of `AML` observations in the middle leaves of the dendrogram, there's no clean way to cut the tree isolating the two cancer types into their own clusters. The smallest number of K that looks to isolate most of the `AML` somewhat well is K$=5$.

```r
plot(hier.cluster, xlab = "", ylab = "", axes = FALSE, main = "", sub = "", lwd = 2)

# Find correct amount of clusters
K <- 5
a <- rect.hclust(hier.cluster, k=K)
```

But even with 3 groups of `ALL` and 2 groups of `AML` we still don't have anything close to as good of models as we have managed to build in the previous exercises. Therefore, we can conclude that hierarchical clustering utilizing an agglomerative method does not build efficient models for this data set.
